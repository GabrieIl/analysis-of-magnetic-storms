```python
import pandas as pd
```

```python
with open('tcm31jan.23m') as f:
    dados = f.readlines()
```

```python
def clean_row(row: str):
    row = row.split(' ')
    last_item = row[-1].replace('\n', '') if '\n' in row[-1] else None
    new_row = [i for i in row[:-1] if i]
    new_row.append(last_item) if last_item else None
    return new_row
```

```python
header = clean_row(dados[2].replace('DD', 'day').replace('MM', 'month', 1).replace('YYYY', 'year'))
formated_data = {i: [] for i in header}

def format_data_to_df(formated_data: dict, data: list):
    for i, k in enumerate(formated_data):
        formated_data[k].append(data[i])
    return formated_data
```

```python
[format_data_to_df(formated_data, clean_row(current_row)) for current_row in dados[4:] if current_row]
```

```python
df = pd.DataFrame(formated_data)
```
