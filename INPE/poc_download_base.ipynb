```python
import os
import requests
from bs4 import BeautifulSoup
BASE_URL = 'https://embracedata.inpe.br/'
```

```python

def find_files(url):
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, "html.parser")
    return [a['href'] for a in soup.find_all('a', href=True) if not ';' in a['href'] and not a['href'].endswith('/')]


def build_url(root_url):
    urls = []
    res = requests.get(root_url)
    soup = BeautifulSoup(res.text, "html.parser")
    relative_paths = [a['href'] for a in soup.find_all('a', href=True) if len(a['href']) > 1 and a['href'][-1] == '/']
    
    if relative_paths:
        for path in relative_paths:
            if path not in root_url:
                urls.append(root_url + path)
    return urls
```

```python
magnetometer_url = [build_url(i) for i in build_url('https://embracedata.inpe.br/magnetometer/')]
```

```python
def write_file(data: str, file_name: str):
    os.makedirs(os.path.dirname(file_name), exist_ok=True)
    with open(file_name, 'w') as f:
        f.write(data)

def download_file(url: str):
    data = requests.get(url).text
    write_file(data=data, file_name=url.replace(BASE_URL, ''))


def download_all_files(urls: list):
    for url in urls:
        download_file(url)
```

```python
for station_path in magnetometer_url:
    for year_path in station_path:
        url = [year_path + file_name for file_name in find_files(year_path)]
        download_all_files(url)
```
